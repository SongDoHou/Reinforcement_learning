```python
x = np.linspace(-8,8)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def relu(x):
    return np.max(0,x)
```

```python
def Newton_Raphson(x):
    return x-fx(x)/grad(x)
```

# 최적화 그림

```python
import torch
x = np.linspace(-2,2,1000)
y = np.linspace(-3,1,1000)
X,Y = np.meshgrid(x,y)
```

```python
import torch.optim as optim

def loss(x,y):
    return (1+(x+y+1)**2*(19-14*x+3*x**2-14*y+6*x*y+3*y**2))*(30+(2*x-3*y)**2*(18-32*x+12*x**2+48*y-36*x*y+27*y**2))
```

```python
from matplotlib import ticker
Z = loss(X,Y)
plt.contour(X,Y,Z,locator=ticker.LogLocator(),levels=10,colors='white')
plt.contourf(X,Y,Z,locator=ticker.LogLocator(),levels=100)
plt.colorbar()
```

```python
x_tensor = torch.Tensor([1.0])
y_tensor = torch.Tensor([0.5])
print(x_tensor.shape)
print(y_tensor.shape)
cost = loss(x_tensor,y_tensor)
x_tensor.requires_grad=True
```

```python
x_start = -0.5
y_start = -1.5
```

```python
x_tensor = torch.Tensor([x_start])
y_tensor = torch.Tensor([y_start])
x_tensor.requires_grad=True
y_tensor.requires_grad=True
optimSGD = optim.SGD([x_tensor,y_tensor],lr=0.00001)
SGDxy = []
for i in range(100):
    cost = loss(x_tensor,y_tensor)
    optimSGD.zero_grad()
    cost.backward()
    optimSGD.step()
    SGDxy.append([x_tensor.item(),y_tensor.item()])
    if i % 10 == 0:
        print(f'{i} trial')
        print(f'   X: {x_tensor.item():3f}, Y: {y_tensor.item():.3f}, cost: {cost.item()}')
```

```python
x_tensor = torch.Tensor([x_start])
y_tensor = torch.Tensor([y_start])
x_tensor.requires_grad=True
y_tensor.requires_grad=True
optimMomentum = optim.SGD([x_tensor,y_tensor],lr=0.00001,momentum=0.9)
Momentumxy = []
for i in range(100):
    cost = loss(x_tensor,y_tensor)
    optimMomentum.zero_grad()
    cost.backward()
    optimMomentum.step()
    Momentumxy.append([x_tensor.item(),y_tensor.item()])
    if i % 10 == 0:
        print(f'{i} trial')
        print(f'   X: {x_tensor.item():3f}, Y: {y_tensor.item():.3f}, cost: {cost.item()}')
```

## 확률통계 그림

```python
x = np.linspace(-3,3,50)

def Gaussian(x,mean=0,std=1):
    return 1/(np.sqrt(2*np.pi*std**2))*np.exp(-((x-mean)/(2*std**2))**2)
```

```python
stds = [2,1,0.5,0.3] #[0.3, 0.5, 1, 2]
#lines = ['-*','--o','-.',':^']
lines = ['--*','-->','--o','--^']
for idx,std in enumerate(stds):
    y = Gaussian(x,0,std)
    plt.plot(x,y,lines[idx],label=f'# of posterior:{idx+1}',markersize=6)
    
plt.grid()
plt.legend()
plt.title("Posterior probability update")
```

# 포아송분포

```python
poisson_dict = {}

def poisson_dist(x,lam):
    global poisson_dict
    key = (x,lam)
    if key not in poisson_dict.keys():
        poisson_dict[key] = lam**x*np.exp(-lam)/np.math.factorial(x)
    return poisson_dict[key]
```

```python
lam1 = 1
lam3 = 3
lam5 = 5
x = np.arange(0,11)
x
```

```python
for idx,lam in enumerate([lam1,lam3,lam5]):
    p_dist = []
    for x_ele in x:
        dummy = poisson_dist(x_ele,lam)
        p_dist.append(dummy)
    if idx == 0:
        plt.plot(p_dist,'o-',label='lambda_1')
    elif idx == 1:
        plt.plot(p_dist,'^-',label='lambda_3')
    else:
        plt.plot(p_dist,'*-',label='lambda_5')
plt.legend()
plt.grid()
```
