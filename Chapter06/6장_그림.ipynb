```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
#x = np.linspace(mu - 10*sigma, mu + 3*sigma, 100)
x = np.linspace(-4,4,1000)
plt.plot(x, stats.norm.pdf(x, mu, sigma),label='PDF_1')

mu2 = 0.5
sigma2 = 0.95**0.5
plt.plot(x, stats.norm.pdf(x, mu2, sigma2),'-.',label='PDF_2')

mu3 = -2
sigma3 = 0.5**0.5
plt.plot(x, stats.norm.pdf(x, mu3, sigma3),'--',label='PDF_3')

#plt.grid()
#plt.legend()
plt.show()
```
