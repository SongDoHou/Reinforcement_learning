```python
num_points = 10
x_rand = np.random.uniform(0,1,num_points)
y_rand = np.random.uniform(0,1,num_points)
```

```python
def plot_montecarlo_circle(num_points=10,grid=True,title=True):
    plt.figure(figsize=(10,10))
    plt.plot(x,y_circle,'-k')
    x_rand = np.random.uniform(0,1,num_points)
    y_rand = np.random.uniform(0,1,num_points)
    
    in_circle,outer_circle = [],[]
    for idx in range(num_points):
        if y_rand[idx] <= np.sqrt(1-x_rand[idx]**2):
            in_circle.append([x_rand[idx],y_rand[idx]])
        else:
            outer_circle.append([x_rand[idx],y_rand[idx]])
    in_circle = np.array(in_circle)
    outer_circle = np.array(outer_circle)
    
    plt.plot(in_circle[:,0],in_circle[:,1],'o',markersize=8)
    plt.plot(outer_circle[:,0],outer_circle[:,1],'*',markersize=8)
    
    pi_value = 4*len(in_circle)/num_points
    print(f'Total num_points: {num_points}')
    print(f'The number of in circles: {len(in_circle)}')
    print(f'Monte Carlo estimated pi: {pi_value}')
    
    if grid:
        plt.grid()
    if title:
        plt.title(f'Estimated the value of pi: {pi_value}')
```

```python
plot_montecarlo_circle(10)
```

```python
plot_montecarlo_circle(100)
```

```python
plot_montecarlo_circle(1000)
```

```python
plot_montecarlo_circle(10000)
```

```python
plot_montecarlo_circle(1000000)
```

```python
# eligibility traces
# entire point 200
# shock point - 10 20 30, 50 80
gamma = 0.95
lam = 0.95
ts = np.linspace(0,199,200)
shock = [10,15,20,60,70,120]
E_trace = np.zeros_like(ts)
for t,_ in enumerate(ts):
    if len(shock) != 0 and t == shock[0]:
        shock.pop(0)
        E_trace[t] = gamma*lam*E_trace[t-1]+1
    elif t >= 1:
        E_trace[t] = gamma*lam*E_trace[t-1]
```

```python
plt.plot(ts,E_trace)
```
