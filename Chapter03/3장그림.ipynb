```python
import graphviz
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import Image
```

```python
dot = graphviz.Digraph('round-table', comment='The Round Table')  
dot.node('A', '저축(S1)',pos='50,50')  
dot.node('B', '소비(S2)',pos='0,0')
dot.node('L', '투자(S3)',pos='100,0')

dot.edges(['AB', 'AL','BL'])
dot.edges(['BA', 'LA','LB'])
dot.edges(['AA', 'BB','LL'])
dot.format='png'
```

```python
print(dot.source)
```

```python
dot.render(directory='./', view=False)  
```
